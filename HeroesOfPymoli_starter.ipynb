```python
# Create bins in which to place values based upon TED Talk views
bins = [0, 9, 14, 19, 24, 29, 34, 39, 119]

# Create labels for these bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data["Age_groups"] = pd.cut(purchase_data["Age"], bins, labels=group_names)
purchase_data.head()
```

```python
totalCount = purchase_data.groupby(["SN","Age_groups"]).count()["Purchase ID"].groupby(["Age_groups"]).count()
totalCount
```
